# Q2. 범죄율은 전체적으로 증가하는 양상을 보이고 있다. 그럼 사건을 담당해야 할 경찰들의 수도 증가하고 있는가?

```python
police = pd.read_csv("police.csv", encoding="cp949").set_index("연도")
police = police.transpose()
```

```python
plt.plot(year, police["십만 명당 경찰관 수"])
plt.title("경찰관 수의 변화")
plt.show()
```

```python
police["한 범죄를 맡는 경찰의 수"] = round(police["십만 명당 경찰관 수"] / crime["전체"], 2)
plt.plot(year, police["한 범죄를 맡는 경찰의 수"])
plt.title("범죄 하나에 배정되는 경찰관 수")
plt.show()
```

# Q3. 성폭행 범죄의 비율은 어떻게 변하고 있는가?

```python
sex_offense = crime.loc[:, ["성폭행", "전체"]]
sex_offense["성폭행_비율"] = round(sex_offense["성폭행"] / sex_offense["전체"], 3)
plt.plot(year, sex_offense["성폭행_비율"])
plt.title("전체 범죄 대비 성폭행 범죄의 비율")
plt.show()
```
